## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import numpy as np

#  Import and read the charity_data.csv.
import pandas as pd 
app_df = pd.read_csv("../Resources/charity_data.csv")
app_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
app_df = app_df.drop(['EIN', 'NAME'], axis = 1)
app_df
```

```python
# Determine the number of unique values in each column.
app_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = []
for x in app_df['APPLICATION_TYPE'].unique():
    if len(app_df.loc[app_df['APPLICATION_TYPE'] == x]) < 500:
        application_types_to_replace.append(x)

# Replace in dataframe
for app in application_types_to_replace:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
app_df['CLASSIFICATION'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
for x in app_df['CLASSIFICATION'].unique():
    if len(app_df.loc[app_df['CLASSIFICATION'] == x]) < 1880:
        classifications_to_replace.append(x)

# Replace in dataframe
for cls in classifications_to_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_df = pd.get_dummies(app_df)
app_df
```

```python
# Split our preprocessed data into our features and target arrays
y = app_df['IS_SUCCESSFUL'].values
X = app_df.drop(["IS_SUCCESSFUL"], axis = 1).values

# Split the preprocessed data into a training and testing dataset
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(Xtrain)

# Scale the data
X_train_scaled = X_scaler.transform(Xtrain)
X_test_scaled = X_scaler.transform(Xtest)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
numinputfeatures = len(Xtrain[0])
hiddennodes1 = 80
hiddennodes2 = 30
numinputfeatures

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units = hiddennodes1,
                          input_dim = numinputfeatures,
                          activation = "relu")
)

# Second hidden layer

nn.add(tf.keras.layers.Dense(units = hiddennodes2,
                             activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1,
                             activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = "binary_crossentropy",
           optimizer = "adam",
           metrics = ["accuracy"])
```

```python
# Train the model
fitmodel = nn.fit(X_train_scaled,
                  ytrain,
                  epochs = 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,
                                         ytest,
                                         verbose = 2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```
